```python
from sklearn.preprocessing import OneHotEncoder, FunctionTransformer, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import StratifiedShuffleSplit, cross_validate

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB

from interpret.glassbox import ExplainableBoostingClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import roc_auc_score, make_scorer
from sklearn.impute import SimpleImputer

def format_n(x):
    return "{0:.3f}".format(x)

def process_model(clf, name, X, y, n_splits=3):
    # Evaluate model
    ss = StratifiedShuffleSplit(n_splits=n_splits, test_size=0.25, random_state=1337)
    y_series = pd.Series(y)
    num_classes = y_series.nunique()
    
    if num_classes == 2:
        scores = cross_validate(
        clf, X, y, scoring='roc_auc', cv=ss,
        n_jobs=None, return_estimator=True)
    elif num_classes > 2:
        multi_roc_scorer = make_scorer(lambda y_in, y_p_in: roc_auc_score(y_in, y_p_in, multi_class='ovr'), needs_proba=True)
        scores = cross_validate(
        clf, X, y, scoring=multi_roc_scorer, cv=ss,
        n_jobs=None, return_estimator=True)
    record = dict()
    record['model_name'] = name
    record['fit_time_mean'] = format_n(np.mean(scores['fit_time']))
    record['fit_time_std'] = format_n(np.std(scores['fit_time']))
    record['test_score_mean'] = format_n(np.mean(scores['test_score']))
    record['test_score_std'] = format_n(np.std(scores['test_score']))

    return record



def benchmark_models(dataset_name, X, y, ct=None, n_splits=3, random_state=1337):
    if ct is None:
        is_cat = np.array([dt.kind == 'O' for dt in X.dtypes])
        cat_cols = X.columns.values[is_cat]
        num_cols = X.columns.values[~is_cat]

        cat_ohe_step = ('ohe', OneHotEncoder(sparse=False,
                                             handle_unknown='ignore'))

        cat_pipe = Pipeline([cat_ohe_step, ('imputer', SimpleImputer(strategy='mean'))])
        num_pipe = Pipeline([('identity', FunctionTransformer()),('imputer', SimpleImputer(strategy='mean'))])
        transformers = [
            ('cat', cat_pipe, cat_cols),
            ('num', num_pipe, num_cols)
        ]
        ct = ColumnTransformer(transformers=transformers)

    records = []
    dataset_records = [] 
    summary_record = {}

    summary_record = {'dataset_name': dataset_name, 'Number of data:': X.shape[0], 'Number of features:': X.shape[1], 'Number of classes to be predicted': y.nunique()}

    dataset_records.append(summary_record)

    print()
    print('-' * 78)
    print(dataset_name)
    print('-' * 78)
    print(summary_record)
    print()
    
    pipe = Pipeline([
        ('ct', ct),
        ('std', StandardScaler()),
        ('lr', LogisticRegression(random_state=random_state)),
    ])
    record = process_model(pipe, 'lr', X, y, n_splits=n_splits)
    print(record)
    record.update(summary_record)
    records.append(record)

    pipe = Pipeline([
        ('ct', ct),
        ('rf-100', RandomForestClassifier(n_estimators=100, n_jobs=-1, random_state=random_state)),
    ])
    record = process_model(pipe, 'rf-100', X, y, n_splits=n_splits)
    print(record)
    record.update(summary_record)
    records.append(record)
    
    pipe = Pipeline([
        ('ct', ct),
        ('xgb', XGBClassifier(random_state=random_state, eval_metric='logloss')),
    ])
    label_encoder = LabelEncoder()
    y_encoded = label_encoder.fit_transform(y)
    record = process_model(pipe, 'xgb', X, y_encoded, n_splits=n_splits)
    print(record)
    record.update(summary_record)
    records.append(record)

    # No pipeline needed due to EBM handling string datatypes
    ebm_inter = ExplainableBoostingClassifier(n_jobs=-1, random_state=random_state)
    record = process_model(ebm_inter, 'ebm', X, y_encoded, n_splits=n_splits)
    print(record)
    record.update(summary_record)
    records.append(record)


    # Decision Tree
    pipe = Pipeline([
        ('ct', ct),
        ('dt', DecisionTreeClassifier(random_state=random_state)),
    ])
    record = process_model(pipe, 'dt', X, y, n_splits=n_splits)
    print(record)
    record.update(summary_record)
    records.append(record)

    # Naive Bayes
    pipe = Pipeline([
        ('ct', ct),
        ('nb', GaussianNB()),
    ])
    record = process_model(pipe, 'nb', X, y, n_splits=n_splits)
    print(record)
    record.update(summary_record)
    records.append(record)

    return records, dataset_records
```

```python
glassbox_models = ['lr', 'ebm', 'dt', 'nb']
blackbox_models = ['xgb', 'rf-100']

record_df['test_score_mean'] = record_df['test_score_mean'].astype(str).str.replace(',', '.').astype(float)


glassbox_df = record_df[record_df['model_name'].isin(glassbox_models)]
blackbox_df = record_df[record_df['model_name'].isin(blackbox_models)]


glassbox_mean_accuracy_per_dataset = glassbox_df.groupby('dataset_name')['test_score_mean'].mean()
blackbox_mean_accuracy_per_dataset = blackbox_df.groupby('dataset_name')['test_score_mean'].mean()


combined_accuracy = pd.DataFrame({
    'glassbox': glassbox_mean_accuracy_per_dataset,
    'blackbox': blackbox_mean_accuracy_per_dataset
}).fillna(0)


accuracy_comparison = combined_accuracy['glassbox'] - combined_accuracy['blackbox']


for dataset, difference in accuracy_comparison.items():
    print(f'Dataset: {dataset}')
    if difference > 0:
        print(f'El grupo "glassbox" es {abs(difference):.3f} más preciso en promedio que el grupo "blackbox"')
    elif difference < 0:
        print(f'El grupo "blackbox" es {abs(difference):.3f} más preciso en promedio que el grupo "glassbox"')
    else:
        print('Ambos grupos tienen la misma precisión en promedio')
    print()
```

```python
def find_best_model_per_dataset(record_df, glassbox_models, blackbox_models, score_column='test_score_mean'):
    """
    Encuentra el modelo con mejor rendimiento (y su grupo) para cada dataset y devuelve un DataFrame con los resultados.

    Args:
        record_df (pd.DataFrame): DataFrame con registros de rendimiento de modelos (debe incluir nombres de modelos, nombres de datasets y la columna de puntuación).
        glassbox_models (list): Lista de nombres de modelos glassbox.
        blackbox_models (list): Lista de nombres de modelos blackbox.
        score_column (str): Nombre de la columna que contiene las puntuaciones de los modelos.
    
    Returns:
        pd.DataFrame: DataFrame con los mejores modelos para cada dataset y sus respectivas puntuaciones.
    """

    
    record_df[score_column] = (
        record_df[score_column]
        .astype(str)
        .str.replace(',', '.', regex=False)
        .astype(float)
    )

    
    best_models = record_df.loc[record_df.groupby('dataset_name')[score_column].idxmax()]

    
    best_models['model_group'] = np.where(best_models['model_name'].isin(glassbox_models), 'Glassbox', 'Blackbox')

    
    result_df = best_models[['model_group', 'dataset_name', 'model_name', score_column]].reset_index(drop=True)

    return result_df


record_df = pd.read_csv("benchmark_result.csv")
glassbox_models = ['lr', 'ebm', 'dt', 'nb']
blackbox_models = ['xgb', 'rf']
result_df = find_best_model_per_dataset(record_df, glassbox_models, blackbox_models)



print(result_df)


```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import pandas as pd

record_df = pd.read_csv("benchmark_result.csv")
glassbox_models = ['lr', 'ebm', 'dt', 'nb']
blackbox_models = ['xgb', 'rf']
best_models = find_best_model_per_dataset(record_df, glassbox_models, blackbox_models)

if best_models is not None and not best_models.empty:

    color_map = {'Glassbox': 'skyblue', 'Blackbox': 'salmon'}
    colors = best_models['model_group'].map(color_map)

    plt.figure(figsize=(10, 6))
    bars = plt.bar(best_models['dataset_name'], best_models['test_score_mean'].round(1), color=colors)

    for bar, model_name in zip(bars, best_models['model_name']):
        height = bar.get_height()
        label_text = f"{model_name}"
        plt.text(bar.get_x() + bar.get_width() / 2, height, label_text,
                 ha='center', va='bottom', rotation=45)


    plt.xlabel('Conjunto de datos', fontsize=12)
    plt.ylabel('Puntuación media', fontsize=12)
    plt.title('Mejor modelo y grupo para cada conjunto de datos', fontsize=14)
    plt.xticks(rotation=45)

    import matplotlib.patches as mpatches
    glassbox_patch = mpatches.Patch(color='skyblue', label='Modelos transparentes')
    blackbox_patch = mpatches.Patch(color='salmon', label='Modelos caja negra')
    plt.legend(handles=[glassbox_patch, blackbox_patch])
    
    plt.tight_layout()

    plt.show()
else:
    print("No valid model data found. Check your input data or model lists.")
```

```python
all_models = ['lr', 'ebm', 'dt', 'nb', 'xgb', 'rf-100']

model_names = []
mean_accuracies = []

for model in all_models:
    model_df = record_df[record_df['model_name'] == model]
    
    mean_accuracy = model_df['test_score_mean'].astype(float).mean()
    
    model_names.append(model)
    mean_accuracies.append(mean_accuracy * 100)  

results_df = pd.DataFrame({
    'Model': model_names,
    'Mean Accuracy (%)': mean_accuracies
})

print(results_df)

plt.figure(figsize=(10, 6))
colors = ['blue', 'green', 'red', 'purple', 'orange', 'cyan']  # Colores para las barras
bars = plt.bar(results_df['Model'], results_df['Mean Accuracy (%)'], color=colors)

# Crear etiquetas personalizadas
legend_elements = [
    mpatches.Patch(color='blue', label='lr - Logical Regression'),
    mpatches.Patch(color='green', label='ebm - Explainable Boosting Machine'),
    mpatches.Patch(color='red', label='dt - Decision Tree'),
    mpatches.Patch(color='purple', label='nb - Naive Bayes'),
    mpatches.Patch(color='orange', label='xgb - XGBoosting'),
    mpatches.Patch(color='cyan', label='rf-100 - Random Forest')
]

# Agregar la leyenda al gráfico


bars = plt.bar(results_df['Model'], results_df['Mean Accuracy (%)'], color=['blue', 'green', 'red', 'purple', 'orange', 'cyan'])
plt.xlabel('Modelo')
plt.ylabel('Precisión media (%)')
plt.title('Precisión media de cada modelo')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, height, f'{height:.2f}%', ha='center', va='bottom')
plt.legend(handles=legend_elements)
plt.show()

```

```python
record_df['fit_time_mean'] = record_df['fit_time_mean'].astype(float)

ebm_df = record_df[record_df['model_name'] == 'ebm']

for dataset in ebm_df['dataset_name'].unique():

    dataset_df = record_df[record_df['dataset_name'] == dataset]
    
    ebm_fit_time = dataset_df[dataset_df['model_name'] == 'ebm']['fit_time_mean'].values[0]
    
    dataset_df['fit_time_mean'] = dataset_df['fit_time_mean'].astype(float)
    sorted_fit_times = dataset_df.sort_values(by='fit_time_mean', ascending=False)
    
    second_slowest_model = sorted_fit_times.iloc[1]['model_name']
    second_slowest_fit_time = sorted_fit_times.iloc[1]['fit_time_mean']
    
    percentage_diff = ((ebm_fit_time - second_slowest_fit_time) / second_slowest_fit_time) * 100
    
    print(f"Dataset: {dataset}")
    print(f"Modelo EBM vs. {second_slowest_model}: EBM es {percentage_diff:.2f}% más lento")
    print()
```

```python
all_models = ['lr', 'ebm', 'dt', 'nb', 'xgb', 'rf-100']

model_names = []
mean_fit_times = []

for model in all_models:

    model_df = record_df[record_df['model_name'] == model]
    
    mean_fit_time = model_df['fit_time_mean'].astype(float).mean()
    
    model_names.append(model)
    mean_fit_times.append(mean_fit_time)

results_df = pd.DataFrame({
    'Model': model_names,
    'Mean Fit Time': mean_fit_times
})

print(results_df)

plt.figure(figsize=(10, 6))
colors = ['blue', 'green', 'red', 'purple', 'orange', 'cyan']  # Colores para las barras
bars = plt.bar(results_df['Model'], results_df['Mean Fit Time'], color=colors)

# Crear etiquetas personalizadas
legend_elements = [
    mpatches.Patch(color='blue', label='lr - Logical Regression'),
    mpatches.Patch(color='green', label='ebm - Explainable Boosting Machine'),
    mpatches.Patch(color='red', label='dt - Decision Tree'),
    mpatches.Patch(color='purple', label='nb - Naive Bayes'),
    mpatches.Patch(color='orange', label='xgb - XGBoosting'),
    mpatches.Patch(color='cyan', label='rf-100 - Random Forest')
]

# Agregar la leyenda al gráfico
plt.legend(handles=legend_elements)

plt.xlabel('Modelo')
plt.ylabel('Tiempo medio de entrenamiento (segundos)')
plt.title('Tiempo medio de ajuste de cada modelo')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, height, f'{height:.2f} s', ha='center', va='bottom')

plt.show()
```

```python
ebm_df = record_df[record_df['model_name'] == 'ebm']

percentage_diffs = []

for dataset in ebm_df['dataset_name'].unique():

    dataset_df = record_df[record_df['dataset_name'] == dataset]
    
    ebm_fit_time = float(dataset_df[dataset_df['model_name'] == 'ebm']['fit_time_mean'].values[0])
    
    dataset_df['fit_time_mean'] = dataset_df['fit_time_mean'].astype(float)

    sorted_fit_times = dataset_df.sort_values(by='fit_time_mean', ascending=False)
    second_slowest_fit_time = float(sorted_fit_times.iloc[1]['fit_time_mean'])
    
    percentage_diff = ((ebm_fit_time - second_slowest_fit_time) / second_slowest_fit_time)*100 
    percentage_diffs.append(percentage_diff)

average_percentage_diff = sum(percentage_diffs) / len(percentage_diffs)

print(f"El EBM es en promedio {average_percentage_diff:.2f}% más lento que el segundo modelo más lento.")

```
